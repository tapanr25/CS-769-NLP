```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading datasets
finetune_data = pd.read_csv('/content/train_beta_90.csv')
test_data = pd.read_csv('/content/mira_eval_threshold90.csv')
```

```python
finetune_data.head(2)
```

```python
# extracting labels
finetune_labels = finetune_data['binder']
test_labels = test_data['binder']
```

```python
#calculating binding and non-binding counts
ft_true = list(finetune_labels).count(1)
ft_false = list(finetune_labels).count(0)
test_true = list(test_labels).count(1)
test_false = list(test_labels).count(0)

true_list = []
true_list.append(ft_true)
true_list.append(test_true)

false_list = []
false_list.append(ft_false)
false_list.append(test_false)

n=2
r = np.arange(n)

#plot for showing the distribution of datasets
x_labels = ['Train_dataset', 'Test_dataset']
X_axis = np.arange(len(x_labels))
plt.bar(r, true_list, 0.4, label = 'True', color='lawngreen')
plt.bar(r+0.4, false_list, 0.4, label = 'False', color='sandybrown')
plt.xticks(r + 0.5/2,x_labels)
plt.ylabel("Counts")
plt.xlabel("Dataset")
plt.legend()
```

```python
#plot showing the length distribution of CDR3b sequence
lengths = [len(item) for item in finetune_data['CDR3b']]

plt.hist(lengths, bins=10, edgecolor='black',color='orangered')

plt.xlabel("Sequence Length")
plt.ylabel("Frequency")
plt.title("Length Distribution of cdr3b in train dataset")

plt.show()
```

```python
#plot showing the length distribution of peptide sequence
lengths = [len(item) for item in finetune_data['peptide']]

plt.hist(lengths, bins=10, edgecolor='black',color='orangered')

plt.xlabel("Sequence Length")
plt.ylabel("Frequency")
plt.title("Length Distribution of peptide in finetune dataset")

plt.show()
```

```python
#plot showing the length distribution of CDR3b sequence in test dataset
lengths = [len(item) for item in test_data['CDR3b']]

plt.hist(lengths, bins=10, edgecolor='black',color='chocolate')

plt.xlabel("Sequence Length")
plt.ylabel("Frequency")
plt.title("Length Distribution of cdr3b in test dataset")

plt.show()
```

```python
#plot showing the length distribution of peptide sequence in test dataset
lengths = [len(item) for item in test_data['peptide']]

plt.hist(lengths, bins=10, edgecolor='black',color='chocolate')

plt.xlabel("Sequence Length")
plt.ylabel("Frequency")
plt.title("Length Distribution of peptide in test dataset")

plt.show()
```

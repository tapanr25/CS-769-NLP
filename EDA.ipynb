```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
finetune_data = pd.read_csv('/content/drive/MyDrive/courses/Sem1_courses/Advanced_NLP/project/ESM_finetuned_stapler/train-set_full-seq.csv')
test_data = pd.read_csv('/content/drive/MyDrive/courses/Sem1_courses/Advanced_NLP/project/test_data.csv')
```

```python
finetune_data.head(2)
```

```python
finetune_labels = finetune_data['label_true_pair']
test_labels = test_data['label_true_pair']
```

```python
ft_true = list(finetune_labels).count(True)
ft_false = list(finetune_labels).count(False)
test_true = list(test_labels).count(True)
test_false = list(test_labels).count(False)

true_list = []
true_list.append(ft_true)
true_list.append(test_true)

false_list = []
false_list.append(ft_false)
false_list.append(test_false)

n=2
r = np.arange(n)

x_labels = ['Fine-tune_dataset', 'Test_dataset']
X_axis = np.arange(len(x_labels))
plt.bar(r, true_list, 0.4, label = 'True', color='lawngreen')
plt.bar(r+0.4, false_list, 0.4, label = 'False', color='sandybrown')
plt.xticks(r + 0.5/2,x_labels)
plt.ylabel("Counts")
plt.xlabel("Dataset")
plt.legend()
```
